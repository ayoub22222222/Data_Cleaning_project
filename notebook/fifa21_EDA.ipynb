# Height distribution
just for making things clear for people who dont have the ability 
to understand some menings of the columns 

```python

plt.hist(df['Height'], bins=50, color='blue', edgecolor='black')
plt.title('Height distribution')
plt.xlabel('Height distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# removing outliers
filtred_height = df[df['Height'] < 70]
```

```python
# removing outliers
plt.hist(filtred_height['Height'], bins=50, color='blue', edgecolor='black' )
plt.title('filtred height')
plt.xlabel('Height distribution')
plt.ylabel('Frequency')
plt.show()
```
