## Visuaisation and EDA
- now that our data clean we will perform some EDA to extract some usefull insight from our data set

```python
sns.countplot(x ="Team", data= new_df.head(25))
plt.title("top 10 Team distribution")
plt.xlabel("Nationality")  # Add clear x-axis label
plt.ylabel("Count")  # Add clear y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels if there are many categories
plt.show()
```

```python
sns.boxplot(x = "foot", y = "Shot Power", data = new_df)
plt.title("foot distribution against foot")
plt.show()
```

```python
new_df.columns
```

```python
sns.set_style('whitegrid') 
```

```python
test = new_df.drop(columns=['Name', 'Team', 'Nationality', 'foot'])
```

```python
# plotting correlation heatmap
plt.figure(figsize=(12,6))
corr = test.corr()
mask = np.triu(corr)
sns.heatmap(corr, annot= True, mask= mask, fmt = '.2f')
plt.title("correlation between attributes")
plt.show()
```

```python
# some random stuff 
data = {
    "ProductID": [101, 102, 103, 104],
    "SalesAmount": [200, 150, 400, 300],
    "Region": ["North", "South", "West", "East"]
}
df = pd.DataFrame(data)
```

```python
df.set_index('ProductID', inplace=True)
```

```python
df_filtred = df[df['SalesAmount'] > 200]
print(df_filtred)
```

```python
def add_taxes(sales):
    return sales * 1.10
    pass
```

```python
df['SlaesWithTaxes'] = df['SalesAmount'].apply(add_taxes)
```

```python
df
```

```python
df['ProductSummary'] = df.apply(lambda row: f"{row['Region']} - {row['SalesAmount']}", axis=1)
```

```python
df
```

```python
df['Region'] = df['Region'].map(str.upper)
```

```python
df[['SalesAmount', 'SalesWithTaxes']] = df[['SalesAmount', 'SlaesWithTaxes']].applymap(lambda x: round(x, 2))
```
